````python
from selenium import webdriver
````

````python
driver =webdriver.Chrome()
````

````python
### 웹사이트를 크롬드라이버가 제어하는 브라우저에 오픈
url='https://www.melon.com/chart/index.htm'
driver.get(url)
````

````python
html = driver.page_source
html
````

````python
from bs4 import BeautifulSoup 
````

````python
soup = BeautifulSoup(html, 'html.parser')
````

````python
result = soup.select('table > tbody > tr')
result[0]
````

````python
# 1위만 할당
top1 = result[0]
````

````python
result[0].select('a')[2].text
````

````python
## 필요데이터만 추출 
top1.select('a')[2]
````

## BS로만 오류가 난 사이트 테스트

1. 다나와 
- 검색어 입력, 아이디/패스워드 입력 진행하려면
- 셀레니움의 라이브러리 중 common.by가 필요

````python
from selenium.webdriver.common.by import By
````

````python
# 다나와 사이트 
driver.get('https://www.danawa.com/')
````

````python
## 검색어 입력
input_search = driver.find_element(By.ID, 'AKCSearch')
input_search.clear()
input_search.send_keys('HP노트북')
input_search.submit()
````

````python
html = driver.page_source
````

````python
soup2 = BeautifulSoup(html,'html.parser')
````

````python
## 필요데이터 추출 
soup2.select('ul.product_list > li.prod_item')
````

````python
result2 = soup2.select('ul.product_list > li.prod_item')
````

````python
result2[0].select('p.prod_name > a')[0].text
````

````python
result2[0].select('div.spec_list')[0].text
````
